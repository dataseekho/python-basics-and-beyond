## **Lists Are Mutable Sequences**

  - **Lists** are like a basket where you can add, remove, or rearrange items. It's flexible and changeable.

### **Key Features of Lists:**
1. **Ordered**  Items in a list have a specific order.
2. **Mutable**  You can change, add, or remove items after the list is created.
3. **Allow Duplicates**  You can have the same item in a list more than once.

```python
my_list = [1, 2, 3]
my_list[0] = 5
print(my_list)
```

```python
my_list.append(4)
print(my_list)
```

```python
my_list.remove(2)
```

```python
print(my_list)
```

```python
len(my_list)
```

```python
for item in my_list:
    print(item)
```

### **Sets in Python**

A **set** in Python is a collection that is **unordered**, **mutable**, and **does not allow duplicates**. Sets are useful when you want to store unique items and perform operations like checking for common items or combining two sets.

### **Key Features of Sets:**
1. **Unordered**  The items in a set have no specific order.
2. **Mutable**  You can add or remove items from a set.
3. **No Duplicates**  A set cannot have the same item more than once.

### **Common Set Operations**

| **Operation**                    | **Description**                                 | **Example**                                     |
|-----------------------------------|-------------------------------------------------|-------------------------------------------------|
| `my_set.add(x)`                   | Add an item to the set                          | `my_set.add(4)`                                 |
| `my_set.remove(x)`                | Remove an item from the set                     | `my_set.remove(3)`                              |
| `item in my_set`                  | Check if an item exists in the set              | `2 in my_set`                                   |
| `len(my_set)`                     | Get the number of items in the set              | `len(my_set)`                                   |
| `set1.union(set2)` or `set1 | set2` | Combine two sets, removing duplicates           | `set1.union(set2)` or `set1 | set2`            |
| `set1.intersection(set2)` or `set1 & set2` | Get the common items between two sets    | `set1.intersection(set2)` or `set1 & set2`      |
| `set1.difference(set2)` or `set1 - set2` | Items in set1 but not in set2             | `set1.difference(set2)` or `set1 - set2`        |

# **Tuples in Python**

A **tuple** in Python is an **immutable** sequence type that can hold a collection of items. Once you create a tuple, you cannot change its contents (no adding, removing, or modifying items). Tuples are useful for storing related pieces of information together.

### **Key Features of Tuples:**
1. **Immutable**  Once created, you cannot change a tuple (no adding, removing, or modifying items).
2. **Ordered**  The items in a tuple have a specific order and can be accessed by their index.
3. **Allows Duplicates**  A tuple can have the same item more than once.

### **Summary of Common Tuple Operations**

| **Operation**                   | **Description**                                   | **Example**                                     |
|---------------------------------|---------------------------------------------------|-------------------------------------------------|
| `my_tuple[index]`              | Access an item at a specific index               | `my_tuple[1]`                                   |
| `my_tuple[start:end]`          | Get a slice of the tuple                          | `my_tuple[1:3]`                                 |
| `len(my_tuple)`                 | Get the number of items in the tuple             | `len(my_tuple)`                                 |
| `tuple1 + tuple2`              | Concatenate two tuples                            | `tuple1 + tuple2`                               |
| `my_tuple * n`                  | Repeat the tuple `n` times                       | `my_tuple * 3`                                  |
| `item in my_tuple`             | Check if an item exists in the tuple             | `2 in my_tuple`                                 |
| `my_tuple.count(item)`         | Count how many times an item appears              | `my_tuple.count(2)`                             |
| `my_tuple.index(item)`         | Get the index of the first occurrence of an item  | `my_tuple.index(2)`                             |


```python
# Create a tuple
fruits = ("apple", "banana", "cherry")
```

```python
print(fruits[0:2])
```

```python
# Access an item
print(fruits[1])  # Output: banana

# Slicing
print(fruits[0:2])  # Output: ('apple', 'banana')

# Concatenating tuples
citrus_fruits = ("orange", "lemon")
all_fruits = fruits + citrus_fruits
print(all_fruits)  # Output: ('apple', 'banana', 'cherry', 'orange', 'lemon')

# Counting occurrences
print(all_fruits.count("banana"))  # Output: 1

# Finding index
print(all_fruits.index("cherry"))  # Output: 2
```

# **Dictionaries in Python**

A **dictionary** in Python is a collection of key-value pairs. It is a versatile data structure that allows you to store and retrieve data efficiently using keys. Each key in a dictionary is unique, and it is associated with a specific value.

### **Key Features of Dictionaries:**
1. **Unordered**  The items in a dictionary do not have a specific order.
2. **Mutable**  You can change, add, or remove items after the dictionary has been created.
3. **Key-Value Pairs**  Each item is stored as a pair, with a unique key linked to a value.

### **Creating a Dictionary**
To create a dictionary, use curly braces `{}` and separate keys and values with a colon `:`. You can also use the `dict()` function.

### **Summary of Dictionary Operations**

| **Operation**                       | **Description**                                       | **Example**                                  |
|-------------------------------------|-------------------------------------------------------|----------------------------------------------|
| `my_dict[key]`                     | Access the value associated with the key             | `my_dict["name"]`                           |
| `my_dict[key] = value`             | Add a new key-value pair or update an existing key   | `my_dict["age"] = 13`                       |
| `del my_dict[key]`                 | Remove the key-value pair                             | `del my_dict["name"]`                       |
| `my_dict.pop(key)`                 | Remove and return the value associated with the key   | `my_dict.pop("name")`                       |
| `key in my_dict`                   | Check if a key exists in the dictionary               | `"age" in my_dict`                          |
| `my_dict.keys()`                   | Get all keys in the dictionary                        | `my_dict.keys()`                            |
| `my_dict.values()`                 | Get all values in the dictionary                      | `my_dict.values()`                          |
| `my_dict.items()`                  | Get all key-value pairs in the dictionary            | `my_dict.items()`                           |

### **Summary Chart: For Loops vs. While Loops**

| **Feature**         | **For Loop**                                          | **While Loop**                                        |
|---------------------|------------------------------------------------------|------------------------------------------------------|
| **Definition**      | Iterates over a sequence (list, tuple, string, etc.) | Executes as long as a condition is true              |
| **Syntax**          | `for item in sequence:`                              | `while condition:`                                    |
| **Usage**           | Best when you know the number of iterations          | Best when the number of iterations is unknown        |
| **Control Structure**| Controlled by the sequence length or a range        | Controlled by a condition (True/False)               |
| **Termination**     | Automatically stops when the sequence ends           | Must manually change the condition to stop            |
| **Common Operations**| - Loop through items in a list<br>- Loop through characters in a string<br>- Use `range()` for counting | - Check a condition<br>- Use `break` to exit early<br>- Use `continue` to skip an iteration |

```python
# For loop example
fruits = ["apple", "banana", "cherry"]
for var_name in fruits:
    print(var_name)
```

```python
# While loop example
count = 0
while count < 5:
    print(count)
    count = count + 1
```

```python
# Using break in a while loop
count = 0
while count < 10:
    if count == 5:
        break
    print(count)
    count = count + 1
```

```python
# Using continue in a while loop
count = 0
while count < 5:
    count = count + 1
    if count == 3:
      print("Skipping 3")
      continue
    print(count)
```
