# **Object-Oriented Programming (OOP)**

Object-Oriented Programming (OOP) is a programming paradigm that focuses on creating objects that represent real-world entities. These objects contain **attributes** (data) and **methods** (functions) that define their behavior. OOP revolves around four key concepts:

- **Encapsulation**
- **Abstraction**
- **Inheritance**
- **Polymorphism**




## **Key Points of OOP Concepts**:
0. **Classes and Objects**: Classes are blueprints; objects are instances.
1. **Constructors & Destructors**: Special methods for creating and destroying objects.
2. **Encapsulation**: Protect data and expose methods.
3. **Inheritance**: Child classes inherit from parent classes.
4. **Polymorphism**: Different classes can be treated as the same type.
5. **Abstraction**: Hide complexity and show only essential elements.




## **What is a Class?**

A **class** is like a blueprint for a building. It outlines what the building should look like (its properties) and what functions it should perform (its behaviors).


```python
class Car:
    def __init__(self, color, brand):
        self.color = color   # Attribute
        self.brand = brand   # Attribute

    def start_engine(self):  # Method
        print(f"The {self.color} {self.brand}'s engine has started.")
```

## **What is an Object?**
An **object** is an instance of a class. If the class is a blueprint, an object is the actual building made from that blueprint.





```python
my_car = Car("red", "Toyota")
my_car.start_engine()
```

## **1. Constructors and Destructors**

### **Constructor**
A constructor (`__init__`) is a special method that initializes new objects.


```python
class Car:
    def __init__(self, color, brand):
        self.color = color
        self.brand = brand
        print(f"The {self.color} {self.brand}.")

my_car = Car("red", "Toyota")
```


### **Destructor**
A destructor (`__del__`) is called when an object is destroyed, freeing up resources.

```python
class Car:
    def __init__(self, color, brand):
        self.color = color
        self.brand = brand

    def __del__(self):
        print(f"The {self.color} {self.brand} is being destroyed.")

my_car = Car("red", "Toyota")
del my_car
```

### **Public Attributes**
- Accessible from outside the class. They can be read and modified by any code that has access to the object.

```python
class Student:
    def __init__(self, name, age):
        self.name = name  # Public attribute
        self.age = age    # Public attribute

# Creating a Student object
student1 = Student("Alice", 20)
```

```python
# Accessing public attributes
print(student1.name)  # Output: Alice
print(student1.age)   # Output: 20
```

```python
# Modifying public attributes
student1.age = 21
print(student1.age)   # Output: 21
```

### **Private Attributes -> (Encapsulation)**
- Inaccessible from outside the class. They can only be accessed and modified from within the class itself.

## **1. Encapsulation**

### **What is Encapsulation?**
- **Encapsulation** is a way to keep data safe and organized in a class. It combines data (attributes) and the actions that can be performed on that data (methods) into a single unit (the class).

### **Why Use Encapsulation?**
- **Data Protection**: It prevents accidental changes to important information.
- **Control Access**: It allows you to control how the data is accessed or modified.
  

```python
class Student:
    def __init__(self, name, age):
        self.__name = name  # Private attribute
        self.__age = age    # Private attribute

    def get_name(self):  # Public method to access the private name attribute
        return self.__name

    def get_age(self):  # Public method to access the private age attribute
        return self.__age

# Creating a Student object
student1 = Student("Bob", 22)
```

```python
# Attempting to access private attributes directly (will raise an AttributeError)
print(student1.__name)
```

```python
# Accessing private attributes through public methods
print(student1.get_name())
print(student1.get_age())
```

```python
# Base class
class Course:
    def __init__(self, course_fee,course_name, credits, instructor):
        self.__course_fee = course_fee    # Private attribute
        self.course_name = course_name    # Public attribute
        self.credits = credits            # Public attribute
        self.instructor = instructor      # Public attribute

    # Public method to get the course name
    def get_course_name(self):
        return self.course_name

    # Public method to get credits
    def get_credits(self):
        return self.credits

    # Public method to get instructor name
    def get_instructor(self):
        return self.instructor

    def get_syllabus(self):
        return f"Syllabus for {self.get_course_name()} by {self.get_instructor()}."

    def enroll_student(self, student_name):
        return f"{student_name} has been enrolled in {self.get_course_name()}."


comsats = Course("1000","CS","100","Usama")
comsats.get_course_name()
Nust = Course("1000","Python","100","Ali")
Nust.get_syllabus()
```

## **2. Inheritance**

### **What is Inheritance?**
Inheritance allows a new class (child) to inherit properties and methods from an existing class (parent).


```python
class Parent:
    def __init__(self):
        print("Parent class initialized.")

    def parent_greet(self):
        print("Hello from the Parent class.")

class Child(Parent):
    def __init__(self):
        super().__init__()  # Calls the Parent's __init__ method
        print("Child class initialized.")

    def greet(self):
        super().parent_greet()  # Calls the Parent's greet method
        print("Hello from the Child class.")

# Creating an instance of the Child class
child_instance = Child()
child_instance.greet()
```


## **3. Polymorphism**

### **What is Polymorphism?**
Polymorphism means that different classes can be treated as instances of the same class through a common interface.


### **Polymorphism: Different Behaviors**
Polymorphism allows different types of courses to have a method that behaves differently based on the course type.

- Both **ProgrammingCourse** and **MathematicsCourse** can have a method called `get_schedule()`. However, the output will be different:
  - **ProgrammingCourse**: Might include coding assignments and lab sessions.
  - **MathematicsCourse**: Might include theory lectures and problem-solving sessions.

```python
# Define a Dog class
class Dog:
    def speak(self):
        return "Woof!"

# Define a Cat class
class Cat:
    def speak(self):
        return "Meow!"

# Create instances of Dog and Cat
my_dog = Dog()
my_cat = Cat()
```

```python
# Function to demonstrate polymorphism
def animal_sound(animal):
    print(animal.speak())
```

```python
Class -> Dog -> speak()
Class -> Cat -> speak()

class_dog = Dog()

def fun_name(varName):
    varName.speak()


fun_name(class_dog)
```

```python
# Call the function with different animals
animal_sound(my_dog)
animal_sound(my_cat)
```

## **4. Abstraction**

### **What is Abstraction?**
Abstraction hides the complex details and shows only the essential features.

### **Abstraction: Hiding Complexity**
Abstraction allows you to interact with a course without needing to understand all the underlying details.

- You can enroll in a course by simply calling the `enroll_student()` method without needing to know how the enrollment process works behind the scenes (like database operations, checks, etc.).

**ABC stands for Abstract Base Class.**
It is a class that helps define an abstract class in Python

```python
from abc import ABC, abstractmethod

# Abstract class
class Student(ABC):
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id

    @abstractmethod
    def display_info(self):
        pass

# Concrete class for students
class RegularStudent(Student):
    def display_info(self):
        print(f"Student Name: {self.name}, Student ID: {self.student_id}")

# Example usage
def main():
    # Create a RegularStudent instance
    student = RegularStudent("John Doe", "S001")

    # Display student information
    student.display_info()

if __name__ == "__main__":
    main()
```
