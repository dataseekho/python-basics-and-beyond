# Error Handling in Python:

Error handling is a way to respond to errors gracefully, ensuring your program doesn’t crash unexpectedly. In Python, this is done using try, except, and finally blocks.

1. **Try Block**:
   - **Key Point**: It’s like taking a chance on something that might fail.
   - **Example**: Trying to divide a number.

2. **Except Block**:
   - **Key Point**: It’s like having a backup plan if things go wrong.
   - **Example**: If you try to divide by zero, it catches that mistake.

3. **Else Block**:
   - **Key Point**: It’s a happy path—what to do if everything works out!
   - **Example**: If you divide successfully, print the result.

4. **Finally Block**:
   - **Key Point**: This is for the “cleanup crew”—it always runs at the end.
   - **Example**: Closing a notebook after you’re done, no matter what happened.

5. **Raising Exceptions**:
   - **Key Point**: This is like telling someone they made a mistake on purpose.
   - **Example**: If someone enters a negative number, you tell them it’s not allowed.

```python
def divide_numbers(x, y):
    try:
        result = x / y  # Try block
    except ZeroDivisionError:  # Except block
        print("Error: Cannot divide by zero!")
    else:  # Else block
        print(f"Result: {result}")
    finally:  # Finally block
        print("Execution complete.")

def get_positive_number(number):
    if number <= 0:  # Raising exceptions
        raise ValueError("Input must be a positive number.")

try:
    divide_numbers(10, 1)  # Test with a zero denominator
    get_positive_number(5)
except ValueError as e:
    print(e)
```

# **2. File Handling**

### **Definition:**  
File handling in Python allows you to read from and write to files. This is essential for data storage, retrieval, and processing.

#### **4 Basic Operations:**
- **Open a file:** `open(filename, mode)`
- **Read from a file:** `file.read()`
- **Write to a file:** `file.write(data)`
- **Close a file:** `file.close()`


#### 1. **Open a File:**
   - **Syntax:** `open(filename, mode)`
   - **Parameters:**
     - `filename`: The name of the file to open.
     - `mode`: The mode in which the file should be opened (e.g., read, write).
   - **Modes:**
     - `'r'`: Read (default). Opens a file for reading.
     - `'w'`: Write. Opens a file for writing (overwrites if it exists).
     - `'a'`: Append. Opens a file for appending (creates a new file if it doesn't exist).
#### 2. **Read from a File:**
   - **Methods:**
     - `file.read(size)`: Reads the entire file or a specific number of bytes.
     - `file.readline()`: Reads the next line from the file.
     - `file.readlines()`: Reads all the lines in a file and returns them as a list.

#### 3. **Write to a File:**
   - **Method:** `file.write(data)`


#### 4. **Close a File:**
   - **Method:** `file.close()`
   - **Purpose:** To close the file and free up system resources.

```python
print("Data Seekho\n")
print("Session")
```

```python
# Open a file in read mode
file = open('example.txt', 'r')
content = file.read()  # Read entire file content
print(content)
```

```python
# Open a file in write mode
file = open('example.txt', 'w')
file.write('Hello, World!\n')  # Write to the file
file.write('This is a new line.\n')
file.close()  # Close the file
```

```python
# Open a file in write mode
file = open('example.txt', 'a')
file.write('Hello, World!\n')  # Write to the file
file.write('This is a new line.\n')
file.close()  # Close the file
```

# **Using with Statement**

A better practice in file handling is to use the with statement. This ensures that the file is automatically closed after the block is executed, even if an error occurs.



```python
# Open a file in read mode
file = open('example.txt', 'r')
content = file.read()  # Read entire file content
print(content)
```

```python
# Using 'with' to handle files
with open('example.txt', 'w') as example:
    example.write('Hello, World!\n')  # Write to the file

# The file is automatically closed after this block

# Reading the file
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)  # Output the content
```
