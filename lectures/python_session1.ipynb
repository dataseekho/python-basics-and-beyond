### **Type of Errors in Python**

In Python, errors can be classified into three main categories: **Syntax Errors**, **Runtime Errors**, and **Logical Errors**

### **1. Syntax Errors**
These errors occur when the Python interpreter encounters incorrect syntax in the code. This prevents the code from being executed.

````python
print("Welcome to Data Seekho! Python Class")
````

````python
print("Welcome to Data Seekho! Python Class"
````

### **2. Runtime Errors**
These errors occur during the execution of the program. They can happen for various reasons, such as invalid operations or references.

````python
# Division by zero
result = 10 / 0
print(result)
````

````python
# Using a variable that has not been defined
a = 1
print(a)
````

### **3. Logical Errors**
These errors occur when the program runs without crashing but produces incorrect results due to flaws in the logic of the code. These can be the hardest to identify.

````python
# Incorrect calculation of the area of a rectangle

Length  = 10
Width  = 2
area_of_rectangle = Length * Width
print(area_of_rectangle)
````
